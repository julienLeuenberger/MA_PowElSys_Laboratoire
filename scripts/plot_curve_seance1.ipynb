```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Records
ta = 3, fp = 5k => T0002.CSV \
ta = 3, fp = 16k => T0003.CSV \
ta = 4.5, fp = 5k => T0001.CSV \
ta = 4.5, fp = 16k => T0000.CSV \
ta 4.5 devient ta 5 // erreur dans les mesures \
Hypothèses :
- fp plus haut amène un meilleur suivit de consigne sinus, donc on prend 16kHz 3 et 4.5 ta et on remarque qu'un ta plus lent et meilleur et permet un courant plus élévé
- ensuite on prend ta 3 car c'est le meilleur et on remarque qu'avec fp 16k le sinus est mieux défini

```python
zoom = False
```

```python
start = 1000
end = 25000
ta = 3
fp = 5000
df = pd.read_csv("../mesures/seance_1/T0002ALL.CSV", skiprows=15)
df
u_L3 = df["CH1"].to_numpy()
i_L3 = df["CH2"].to_numpy()
i_L2 = df["CH3"].to_numpy()
time = df["TIME"].to_numpy()
if zoom:
    u_L3 = u_L3[start:end]
    i_L3 = i_L3[start:end]
    time = time[start:end]
plt.figure(figsize=(10, 6), dpi=80)
fig,ax = plt.subplots()
ax.plot(time, u_L3, label=f'u_L3', color='red')
ax.set_xlabel("Time [s]")
ax.set_ylabel("Tension L3 [V]", color='red')
ax2 = ax.twinx()
ax2.plot(time, i_L3, label=f'i_L3', color='blue')
ax2.set_ylabel("Courant L3 [A]", color='blue')
plt.grid()
plt.title(f"Tension et courant L3, ta = {ta} us, fp = {fp/1000} kHz")
te = "zoom" if zoom else "full"
plt.savefig(f"../images/ul3_i3_{str(ta).replace('.', '_')}_{str(fp/1000).replace('.', '_')}kHz_{te}.png")
```
